### 2016 & 2017 Cizgisel Grafigi

<a id= "3_a"></a><br>
##### Grafik icin verinin hazirlanmasi

```python
weeks = LMTF.MerchYearWeek.iloc[0:52].astype(str)
weeks = weeks.apply(lambda x:x[-2:])
```

```python
# 2016 ve 2017 deki "Toplam Sales" degerleri ayri ayri cekildi
Sales_2016 = LMTF[LMTF["MerchYearWeek"].astype(str).apply(lambda x:x[:-2]) == "2016" ]["Toplam Sales"].reset_index()
Sales_2017 = LMTF[LMTF["MerchYearWeek"].astype(str).apply(lambda x:x[:-2]) == "2017" ]["Toplam Sales"].reset_index()

data = {"weeks":(weeks),
        "2016":(Sales_2016["Toplam Sales"]),
        "2017":(Sales_2017["Toplam Sales"])}
data  = pd.DataFrame(data)
```

```python
#data["2016"] = data["2016"]/max(data["2016"])
#data["2017"] = data["2017"]/max(data["2017"])
#data
```

<a id="3_b"><a/><br>
##### Grafik

```python
custom_lines = [Line2D([0], [0], color="r"),
                Line2D([0], [0], color="b")]

plt.subplots(figsize = (25,5))
sns.set(style="darkgrid")
sns.pointplot(x = weeks, y = "2016" ,data = data ,color = "r")
sns.pointplot(x = weeks, y = "2017" ,data = data ,color = "b")
plt.title("2016 & 2017")
plt.xlabel("Weeks")
plt.ylabel("Toplam Sales")
plt.legend(custom_lines, ['2016', '2017'])
plt.show()
```

### 2016 & 2017 Histogram Grafigi

<a id= "4_a"></a><br>
##### Grafik icin verinin hazirlanmasi

```python
new_index = (LMTF["Toplam Sales"].sort_values()).index.values
sorted_data  = LMTF.reindex(new_index)
```

```python
min_sales = min(sorted_data["Toplam Sales"])
max_sales  = max(sorted_data["Toplam Sales"])
data_openness = max_sales - min_sales
group_width = round((data_openness / 21) + 1) # (data_opennes / number of groups) = next natural number
```

```python
X_axis = []
counter  = 0
temp = min_sales
while(counter < 21):
    high_temp = temp + group_width
    X_axis.append(str(temp)+" - "+str(high_temp))
    temp = high_temp+1
    counter += 1 
```

```python
dict_data = {}
for i in range(len(X_axis)):
    dict_data.update({X_axis[i]: len(sorted_data[(sorted_data["Toplam Sales"]  >= int(X_axis[i][0:4])) &  (sorted_data["Toplam Sales"] <= int(X_axis[i][-4:]))]["Toplam Sales"])})
```

```python
index = list(dict_data.keys())
values = list(dict_data.values())
hist_data = pd.DataFrame(dict_data, index=index,columns=['Values'])
hist_data["Values"] = values
```

```python
hist_data.drop(hist_data[hist_data["Values"] == 0].index,inplace=True)
```

<a id= "4_b"></a><br>
##### Grafik 

```python
plt.subplots(figsize = (20,9))
sns.barplot(hist_data.index,hist_data.Values)
plt.title("2016 & 2017")
plt.xlabel("Sales")
plt.ylabel("Number of Weeks")
plt.xticks(rotation=90)
plt.show()
```
